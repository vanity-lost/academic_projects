1.1. Prove (5 pts)
$$ p(x,y|z) = p(x|z)p(y|x,z) $$

1.2. Prove (5 pts)
$$ p(x|y,z) = \frac{p(y|x,z)p(x|z)}{p(y|z)} $$

### Q1 Answer
1.1.
By definition of conditional probability:
$$ p(x|z)p(y|x,z) = \frac{p(x,z)}{p(z)}\cdot\frac{p(x,y,z)}{p(x,z)} $$
$$ p(x|z)p(y|x,z) = \frac{p(x,y,z)}{p(z)} $$
$$ p(x|z)p(y|x,z) = p(x,y|z)$$

Therefore, 
$$ p(x,y|z) = p(x|z)p(y|x,z) $$

Proved.

1.2.
By definition of conditional probability:
$$ \frac{p(y|x,z)p(x|z)}{p(y|z)} = \frac{\frac{p(x,y,z)}{p(x,z)}\cdot\frac{p(x,z)}{p(z)}}{\frac{p(y,z)}{p(z)}} $$
$$ \frac{p(y|x,z)p(x|z)}{p(y|z)} = \frac{p(x,y,z)}{p(y,z)} $$
$$ \frac{p(y|x,z)p(x|z)}{p(y|z)} = p(x|y,z) $$

Therefore, 
$$ p(x|y,z) = \frac{p(y|x,z)p(x|z)}{p(y|z)} $$

Proved.

### Q2 Answer - 2 (Reasoning by showing logic gap):
2.1. Reasoning by showing logic gap: \
Since we have $a \perp b \wedge b \perp c $, we have $p(b,c)=p(b)p(c)$ and $p(a|b,c)=p(a|c)$. Thus, we have
$$p(a,b,c)=p(a|b,c)p(b,c)=p(a|c)p(b)p(c)$$
If we know that $p(a,b,c)=p(a)p(b)p(c)$, we have
\begin{align} 
p(a,b,c)&=p(a|c)p(b)p(c)=p(a)p(b)p(c)\\
p(a|c)&=p(a)\\
\frac{p(a,c)}{p(c)}&=p(a)\\
p(a,c)&=p(a)p(c)\\
i.e.\, a \perp c
\end{align}
. However, since we do not know this assumption ($p(a,b,c)=p(a)p(b)p(c)$), there is a logic gap, and thus $a \perp b \wedge b \perp c \nRightarrow a \perp c$.

Example: \
Assume a is the event "Jack go home for dinner" and b is the event "Alex go home for dinner" where Jack and Alex has no relationships and Jack lives in China and Alex lives in India. Obviously, $a \perp b$. 

When c is the event "it will snow tomorrow in US", we know that $a,b,c$ are mutually independent and thus $p(a,b,c)=p(a)p(b)p(c)=p(a|c)p(b)p(c)$. We have $p(a,c)=p(a)p(c)$, which is $a \perp c$.

However, this is not always correct. Let's change c to be the event "the mother of A require A go home for dinner" and we know $a,b,c$ is not mutually independent and $p(a,b,c)\neq p(a)p(b)p(c)$. Here $p(a,b,c)=p(a|c)p(b)p(c)>p(a)p(b)p(c)$, and thus $p(a|c)>p(a)$, which means $a \perp b \wedge b \perp c \nRightarrow a \perp c$.


2.2. Reasoning by showing logic gap: \
Since $a|b \perp c$, we have $p(a,b|c)=p(a|c)p(b|c)$. Thus, we can marginalize $p(a,b,c)$ over $c$ to get $p(a,b)$, which is
\begin{align} 
p(a,b) = \sum_c p(a,b,c) 
&= \sum_c p(a,b|c)p(c)  \\
&= \sum_c p(a|c)p(b|c)p(c)  \\
\end{align}

Since we want to show $p(a,b)=p(a)p(b)$, if we know that $a \perp c \wedge b \perp c $, we can prove it by 
\begin{align} 
p(a,b) &= \sum_c p(a|c)p(b|c)p(c)  \\
p(a,b) &= p(a)p(b) \sum_c p(c)  \\
p(a,b) &= p(a)p(b)  \\
\end{align}
. However, since we do not know this assumption ($a \perp c \wedge b \perp c $), there is a logic gap, and thus $a \perp b | c \nRightarrow a \perp b$.

Example:  
Assume a is the event "Jack go home for dinner" and b is the event "Alex go home for dinner" where Jack and Alex has no relationships and Jack lives in China and Alex lives in India.

When c is the event "it will snow tomorrow in US", we know that $a,b \perp c \wedge a \perp c \wedge b \perp c $, and thus we have $$p(a,b)= \sum_c p(a|c)p(b|c)p(c)= p(a)p(b) \sum_c p(c)= p(a)p(b).$$

However, this is not always correct. Let's change c to be the event "the mother of A require A go home for dinner." In this case, we still got $a,b \perp c \wedge b \perp c $ but $a,c$ is not independent, which $p(a|c)>p(a)$. Thus, we have 
\begin{align} 
p(a,b)&= \sum_c p(a|c)p(b|c)p(c)\\
&>\sum_c p(a)p(b)p(c)\\
&= p(a)p(b) \sum_c p(c)\\
&= p(a)p(b).  \\
\end{align}

In other words, $p(a,b)>p(a)p(b)$, which means $a,b$ is not independent. 

Thus, there is a logic gap an results in $a \perp b | c \nRightarrow a \perp b$.

### Q4 Answer
4.1 \
Posterior distribution:
$$
p(\theta|y,n) = \frac{\textsf{Binomial}(y|\theta, n) \textsf{Beta}(\theta|\alpha, \beta)}{p(y|n)} = \textsf{Beta}(\theta | \alpha + y, \beta + n - y)
$$

```python
from scipy import stats

def calc_posterior(n, y, theta, alpha, beta):
    new_alpha = alpha + y
    new_beta = beta + n - y
    return stats.beta.pdf(theta, new_alpha, new_beta)
```

### Q5 Answer
5.1 

Since there is only one observation, the likelihood equation of $x \sim Poisson(\lambda)$ is
$$ p(x=n|\lambda)=\frac{e^{-\lambda}\lambda^{n}}{n!}$$

```python
import numpy as np
import math
import matplotlib.pyplot as plt 

times = [0.53, 0.65, 0.91, 1.19, 1.30, 1.33, 1.90, 2.01, 2.48]
n = len(times)
lamda = n/3

def calc_likelihood(lamda, n=list(range(0,21))):
    likelihoods = []
    for i in n:
        likelihood = 1
        likelihood = (np.exp(-lamda) * (lamda ** i) / math.factorial(i))
        likelihoods.append(likelihood)
    return likelihoods

count = list(range(0,21))
plt.plot(count, calc_likelihood(lamda-1), label = "less") 
plt.plot(count, calc_likelihood(lamda), label = "equal") 
plt.plot(count, calc_likelihood(lamda+1), label = "greater") 
plt.xlabel(r'$n$', fontsize='large') 
plt.ylabel(r'$p(n|\lambda)$', fontsize='large') 
plt.title('likelihood distribution') 
plt.legend() 
plt.show() 
```

Thus, the Gamma distribution is a conjugate prior for the Poisson distribution.

When choosing the $\alpha, \beta$ value of the prior, we consider the mean and variance of $\lambda$ could be. We know that in the prior Gamma distribution $p(\lambda|\alpha, \beta)=Gamma(\alpha,\beta)=\frac{\beta^{\alpha} \lambda^{\alpha-1}e^{-\beta\lambda}}{\Gamma(\alpha)}$, we have $\mu=E(\lambda)=\frac{\alpha}{\beta}$, and $\sigma^2=Var(\lambda)=\frac{\alpha}{\beta^2}$. Let's suppose a slightly different distribution on the true values of the data in this problem. Let's set $\alpha=3, \beta=1$, which results in $\mu=3, \sigma^2=3$, as the prior.

```python
from scipy.stats import gamma
import matplotlib.pyplot as plt 

times = [0.53, 0.65, 0.91, 1.19, 1.30, 1.33, 1.90, 2.01, 2.48]

def calc_posterior(t, lamdas):
    prob = []
    alpha = 3
    beta = 1
    n = len([i for i in times if i <= t])
    print(f'T={t:1.1f}: mean: {(alpha+n)/(beta+t)}, variance={(alpha+n)/((beta+t)**2):1.3f}')
    for lamda in lamdas:
        posterior = gamma.pdf(x=lamda, a=alpha+n, scale=1/(beta+t))
        prob.append(posterior)
    return prob

lamdas = list(range(0,11))
plt.plot(lamdas, calc_posterior(0, lamdas), label = "T=0") 
plt.plot(lamdas, calc_posterior(0.5, lamdas), label = "T=0.5") 
plt.plot(lamdas, calc_posterior(1.5, lamdas), label = "T=1.5") 
plt.xlabel(r'$\lambda$', fontsize='large') 
plt.ylabel(r'$p(\lambda|n,T,\alpha,\beta)$', fontsize='large') 
plt.title('posterior distribution') 
plt.legend() 
plt.show() 
```

I also print the information about each posterior Gamma distribution with T, mean, and variance. 

We can see when T=0, there is no data and it is just a prior, which mean=3 and variance=3, as the "T=0" curve shows.

When T=0.5, although there is still no events happened, we still learn something from the no-events data. Since there is no events happened, if there is no prior, the $\mu_\lambda$ should be very close to 0. However, we have a prior said $\mu_\lambda(prior)=3$. But, the data with no events still decreases our belief about the prior which $\mu_\lambda=3$ to $\mu_\lambda=2$, which causes the curve shift to left. Besides, since no events happened in [0, 0.5], it also decreases the variance, resulted in more concentration around the mean and resulted in a higher peak. 

When T=1.5, we have observed 6 events happened so far. If there is no prior, the $\mu_\lambda$ should be very close to 4. However, since we have a prior said $\mu_\lambda(prior)=3$, the prior decreases our belief on data, which results in $\mu_\lambda=3.6$ and the curve shifts to right. And since this time we have 6 events in a relatively long time, the variance increased a little bit, which make curve spread out slightly.

Let a be the number of rolls that the dice is 1. Let b be the dice we choose to roll. In other words, $a\in \{0,1,2\}$, and $b\in \{"D6","D8","D12"\}$.

```python
import random
random.seed(10)

def produce_results(n=2, dices=[6, 8, 12]):
    dice_index = random.randint(1, len(dices))-1
    rolls = []
    for _ in range(n):
        rolls.append(random.randint(1, dices[dice_index]))
    return (dice_index, rolls.count(1))
```

Let's generate randomly 1000 trials where each trail will choose one dice from three, roll the dice twice, and count the number of dice=1.

```python
# prodce the real-world data
data = []
for i in range(1000):
    data.append(produce_results())

# distribution of the data
print({x:data.count(x) for x in data})
```

Then, we could print the the table of joint probability, marginal probability, and conditional probability. 

In Table 1, the first three row and the first coloumn shows the distribution of p(a,b), which shows the structure the settings. In the Table 1, the last colomn is the marginal probability p(a), by $\sum_{b}{p(a,b)}$, and the last row is the marginal probability p(a), by $\sum_{a}{p(a,b)}$.
```python
# joint probability p(a,b) and marginal probability p(a),p(b)
import pandas as pd

table = []
data_dict = {x:data.count(x) for x in data}
for a in range(3):
    row = []
    for b in range(3):
        row.append(data_dict[(b,a)]/len(data))
    row.append(sum(row))
    table.append(row)
row = [0,0,0,0]
for i in table:
    for j in range(len(row)):
        row[j] += i[j]
table.append(row)
df = pd.DataFrame(table, columns = ['D6', 'D8', 'D12', 'p(a)'], index=['0', '1', '2', 'p(b)'])
print('--- Table 1: joint probability and marginal probability ---')
print(df)
print()

# conditional probability: p(a|b)
a_given_b_table = []
for a in range(3):
    row = []
    for b in range(3):
        row.append(table[a][b]/table[-1][b])
    a_given_b_table.append(row)
df = pd.DataFrame(a_given_b_table, columns = ['p(a|D6)', 'p(a|D8)', 'p(a|D12)'], index=['a=0', 'a=1', 'a=2'])
print('--- Table 2: p(a|b) ---')
print(df)
print()

# conditional probability: p(b|a)
b_given_a_table = []
for a in range(3):
    row = []
    for b in range(3):
        row.append(table[a][b]/table[a][-1])
    b_given_a_table.append(row)
df = pd.DataFrame(b_given_a_table, columns = ['b=D6', 'b=D8', 'b=D12'], index=['p(b|a=0)', 'p(b|a=1)', 'p(b|a=2)'])
print('--- Table 3: p(b|a) ---')
print(df)
```

### E.1

The settings of the problem:

Assume we have a case that we train a AI program to learn from throw coins. In the previous experiments, we used Beta distribution as the prior and trained further on the number of AI throw coins until it got a head (Geometric distribution). However, now we makes a new experiment setting. This time we want to know the probability of k heads in n times (Binomial distribution) and want to use the previous experiments (including previous Beta prior) as the prior knowledge. 

First, we will prove that Beta is a conjugate prior of the Geometric distribution.
From $ posterior \propto likelihood\cdot prior $, we got
\begin{align} 
p(\theta|x)& \propto (x|\theta)p(\theta)   \\
&=Geometric(x|\theta)Beta(\theta|alpha, beta) \\
\end{align}

Let $X=x_1,x_2,...,x_m, X \sim Geometric(\theta)$, we have $$p(x|\theta)=\prod_{i=1}^{m}\theta(1-\theta)^{x_i-1}=\theta^m(1-\theta)^{\sum_ix_i-m}.$$ 
Let $\theta \sim Beta(\alpha,\beta)$, we have $$p(\theta|\alpha, \beta)=\frac{1}{B(\alpha, \beta)} \theta^{\alpha-1}(1-\theta)^{\beta-1}$$ 

Thus, we have
\begin{align} 
p(\theta|x)& \propto Geometric(x|\theta)Beta(\theta|alpha, beta)   \\
&=\theta^m(1-\theta)^{\sum_ix_i-m} \frac{1}{B(\alpha, \beta)} \theta^{\alpha-1}(1-\theta)^{\beta-1} \\
\end{align}

Since $\frac{1}{B(\alpha, \beta)}$ is irrelevant to $\theta$, we can remove it and write as
\begin{align} 
p(\theta|x)& \propto \theta^m(1-\theta)^{\sum_ix_i-m} \theta^{\alpha-1}(1-\theta)^{\beta-1}   \\
&=\theta^{m+\alpha-1} (1-\theta)^{\sum_i x_i+\beta-m-1} \\
&\sim Beta(\alpha+m, \beta+\sum_i x_i-m)=Beta(\alpha', \beta')
\end{align}

Thus, the posterior in the previous experiments is a Beta distribution.

We already know that Beta is a conjugate prior of binomial. Thus, we will use directly the previous posterior (a Beta distribution) as the prior of Binomial likelihood.

Let $Y=y_1,y_2,...,y_q, y \sim Binomial(\theta, y, n),$ we have $$ p(y|\theta,n)=\prod_j {n \choose y_j} \theta^{y_j}(1-\theta)^{n-y_j} $$. Since we know ${n \choose y}$ is irrelevant to $\theta$, we can remove it and write as
\begin{align} 
p(y|\theta,n) &\propto \prod \theta^{y_j}(1-\theta)^{n-y_j} \\
&= \theta^{\sum_j y_j}(1-\theta)^{nq-\sum_j y_j}
\end{align}

Thus, the new posterior will be 
\begin{align} 
p(\theta|y)& \propto Binomial(y|\theta, n) Beta(\theta|\alpha', \beta')   \\
&\propto \theta^{\sum_j y_j}(1-\theta)^{nq-\sum_j y_j} \theta^{m+\alpha-1} (1-\theta)^{\sum_i x_i+\beta-m-1} \\
&= \theta^{\alpha+\sum_j y_j +m-1} (1-\theta)^{\beta+\sum_i x_i+nq-\sum_j y_j-m-1} \\
&\sim Beta(\alpha+\sum_j y_j +m, \beta+\sum_i x_i+nq-\sum_j y_j-m)
\end{align}

Thus, the new posterior is also a Beta distribution. Thus, the Beta-Geometric is also a conjuagte prior of the Binomial distribution. 

In this case, we successfully transfer all knowledge about the previous Geometric distribution and Beta prior on a new Binomial model. 

Here's the code for displaying the knowledge transfer and knowledge correction from the misleading prior.

```python
# Let's display it on data.
from numpy.random import geometric, binomial
from scipy import stats
import numpy as np
np.random.seed(2)

# past Beta prior
alpha = 51
beta = 57

# past Geometric Experiments
m = 100
x = geometric(p=0.8, size=m)

# current Binomial experiment
n = 10
real_theta = 0.53

# currect prior Beta-Geometric distribution
a = alpha+m
b = beta+sum(x)-m
theta = np.linspace(0,1, 1000)
p = stats.beta.pdf(theta, a, b)
plt.plot(theta, p, label='prior')

for q in [1, 10, 100, 500]:
    y = binomial(n, p=real_theta, size=q)
    a = alpha+sum(y)+m
    b = beta+sum(x)+n*q-sum(y)-m
    theta = np.linspace(0,1, 1000)
    p = stats.beta.pdf(theta, a, b)
    plt.plot(theta, p, label='q='+str(q))
plt.legend()
plt.xlim((0.4, 0.9))
plt.xlabel(r'$\theta$', fontsize='large') 
plt.ylabel(r'$p(\theta|y)$', fontsize='large') 
plt.title('Binomial posterior with Beta-Geometric prior') 
plt.show()
```

We can see the current prior (Beta-Geometric) is centered at approximately 0.68, which is the result of the combination of past experiment belief learned from past data (which real value of past experiments is $p(\theta)=0.8$) and the past prior (Beta) knowledge ($\alpha=51,\beta=57$). However, in this case, our real value of $\theta$ in current Binomial experiments is 0.53. Thus, we can see with the increase  of data, the $p(\theta|y)$ corrects the misleading prior and becomes concentrated on real value of $\theta$. Thus, in this model, we successfully transfer the past knowledge about Geometric and Beta distribution to the new Binomial task and further fit for the new Binomial data.

### E.2

Let's play with the normal prior conjugate to normal likelihood. A normal prior is conjugate to a normal likelihood with known $\sigma$.

We know that the likelihood of normal distribution $x_1,x_2,...,x_n \sim N(\theta, \sigma_x^2)$ is $$p(x|\theta)=\prod_{i=1}^{n} \frac{1}{\sqrt{2\pi \sigma_{x}^{2}}} e^{-\frac{(x_i-\theta)^{2}}{2\sigma_{x}^{2}}}$$
And the prior of normal distribution $\theta \sim N(\theta_0, \sigma_\theta^2)$ is $$ p(\theta; \theta_0)=\frac{1}{\sqrt{2\pi \sigma_{\theta}^{2}}} e^{-\frac{(\theta-\theta_0)^{2}}{2\sigma_{\theta}^{2}}} $$

Thus, from $ posterior \propto likelihood\cdot prior $, we got
\begin{align} 
p(\theta|x)& \propto  \prod_{i=1}^{n} \frac{1}{\sqrt{2\pi \sigma_{x}^{2}}} e^{-\frac{(x_i-\theta)^{2}}{2\sigma_{x}^{2}}} \cdot \frac{1}{\sqrt{2\pi \sigma_{\theta}^{2}}} e^{-\frac{(\theta-\theta_0)^{2}}{2\sigma_{\theta}^{2}}}\\
&= \frac{1}{\sqrt{2\pi \sigma_{x}^{2}}}\frac{1}{\sqrt{2\pi \sigma_{\theta}^{2}}} e^{\frac{-\sum_{i=1}^{n}(x_i-\theta)^{2}}{2\sigma_{x}^{2}}} \cdot  e^{-\frac{(\theta-\theta_0)^{2}}{2\sigma_{\theta}^{2}}}\\
\end{align}

Since $\frac{1}{\sqrt{2\pi \sigma_{x}^{2}}}\frac{1}{\sqrt{2\pi \sigma_{\theta}^{2}}}$ is irrelevant to $\theta$, we can remove them and write the equation as
\begin{align} 
p(\theta|x)&\propto e^{\frac{-\sum_{i=1}^{n}(x_i-\theta)^{2}}{2\sigma_{x}^{2}}} \cdot  e^{-\frac{(\theta-\theta_0)^{2}}{2\sigma_{\theta}^{2}}}\\
&= e^{\frac{-\sum_{i=1}^{n}x_i^2+2\theta \sum_{i=1}^{n}x_i-n\theta^2}{2\sigma_{x}^{2}} -\frac{\theta^2-2\theta\theta_0+\theta_0^2}{2\sigma_{\theta}^{2}}}\\
&= e^{-\frac{\sum_{i=1}^{n}x_i^2}{2\sigma_{x}^{2}}-\frac{\theta_0^{2}}{2\sigma_{\theta}^{2}}-\frac{\theta^2}{2}[\frac{1}{\sigma_{\theta}^{2}}+\frac{n}{\sigma_{x}^{2}}]+\theta[\frac{\theta_0}{\sigma_{\theta}^{2}}+\frac{\sum_{i=1}^{n}x_i}{\sigma_x^2}]  }\\
\end{align}

And we know that $-\frac{\sum_{i=1}^{n}x_i^2}{2\sigma_{x}^{2}}-\frac{\theta_0^{2}}{2\sigma_{\theta}^{2}}$ is irrelevant to $\theta$, so we can remove them and write as
\begin{align} 
p(\theta|x)&\propto e^{-\frac{\theta^2}{2}[\frac{1}{\sigma_{\theta}^{2}}+\frac{n}{\sigma_{x}^{2}}]+\theta[\frac{\theta_0}{\sigma_{\theta}^{2}}+\frac{\sum_{i=1}^{n}x_i}{\sigma_x^2}]  }\\
\end{align}

Let $\sigma_\theta'^2=(\frac{1}{\sigma_\theta^2}+\frac{n}{\sigma_x^2})^{-1}$, and $\theta'=\sigma_\theta'^2(\frac{\theta_0}{\sigma_\theta^2}+\frac{\sum_{i=1}^{n}x_i}{\sigma_x^2})$. And, since $-\frac{\theta'^2}{2\sigma_\theta^2}$ is irrelevant to $\theta$, we can add it to the equation and we have
\begin{align} 
p(\theta|x)&\propto e^{ -\frac{\theta^2}{2\sigma_\theta'^2}+\frac{\theta\theta'}{\sigma_\theta'^2}-\frac{\theta'^2}{2\sigma_\theta^2} }\\
&= e^{ -\frac{(\sigma-\sigma')^2}{2\sigma_\theta'^2}} \sim N(\theta', \sigma_\theta'^2)
\end{align}
Thus, we got that normal distribution is the conjugate prior of normal distribution.

```python
# let's plot the graph to show this conjugate prior and posterior.
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import math
from numpy.random import normal
np.random.seed(2)

x = np.linspace(-2, 5, 5000)

theta = 3
sigma_x_squared = 2

theta_0 = 0
sigma_theta_squared = 1

for n in [0, 1, 10, 50, 100]:
    x_data = normal(theta, sigma_x_squared, size=n)
    new_sigma_squrared = 1/(1/sigma_theta_squared+n/sigma_x_squared)
    new_theta = new_sigma_squrared * (theta_0/sigma_theta_squared + sum(x_data)/sigma_x_squared)
    plt.plot(x, stats.norm.pdf(x, new_theta, math.sqrt(new_sigma_squrared)), label='n='+str(n))
plt.legend()
plt.xlim((-1.5, 3.5))
plt.xlabel(r'$\theta$', fontsize='large') 
plt.ylabel(r'$p(\theta|x)$', fontsize='large') 
plt.title('Normal posterior with normal prior') 
plt.show()
```

Application setting: Assume we observe the price of BMW cars is a normal distribution and we find it can be fit to the standard normal distribution $Normal(0,1)$. We also know that the price of Buick cars is also a normal distribution, which is very similar to the price of BMW cars. Thus, we want to use the $Normal(0,1)$ of BMW cars as the prior to predict the Buick cars. (Blackbox information: the price of Buick cars is $Normal(3,2)$.)

As we can see from the graph above, the n=0 curve is the prior, which is a standard $Normal(0,1)$ distribution, with $\theta_0 =0, \sigma_\theta=1$. With the increased size of data, the new posterior learn from the data and correct the misleading prior knowledge bit by bit. We can see since the true data is distributed in $\theta=3$ with $\sigma_x=2$, the normal curve continue shift to right until reaching approximately 3. We also can see that when n=100, the peak or the mean $\theta$ is still not exactly 3. This is because the true data has $\sigma_x=2$ and also because the training data is not large enough.
