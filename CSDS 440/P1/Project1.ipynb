```python
# Research Hypothesis part 1: The computer is good at calculating binary tree. We first thought about turning the decision tree to a binary tree to limit the size of the tree where it allows for faster evaluation. We can do this by turning every attribute into a continuous attribute after the ExampleSet is turned into 2D list.

from mainR import getAccuracyAndSize
from matplotlib import pyplot as plt

avgSizeArrO = []      # Origial avgSize
avgSizeArrR = []      # Research avgSize
avgAccuracyArrO = []      # Origial avgAccuracy
avgAccuracyArrR = []      # Research avgAccuracy

depths = [1, 2, 4, 8, 16, 32]
for depth in depths:
    avgAccuracy, avgSize = getAccuracyAndSize('/voting', 1, depth, 0, 0)      # the fifth option is the useBinaryTree option. 0 means model as original. 1 means model the whole tree as a binary tree.
    avgSizeArrO.append(avgSize)
    avgAccuracyArrO.append(avgAccuracy)
    avgAccuracy, avgSize = getAccuracyAndSize('/voting', 1, depth, 0, 1)
    avgSizeArrR.append(avgSize)
    avgAccuracyArrR.append(avgAccuracy)

plt.plot(depths, avgAccuracyArrO, linestyle='dashed', label = 'Original')
plt.plot(depths, avgAccuracyArrR, label = 'Research')
plt.title('Depth-Accuracy')
plt.xlabel('Depth')
plt.ylabel('Accuracy')
plt.legend()
plt.show()

plt.plot(depths, avgSizeArrO, linestyle='dashed', label = 'Original')
plt.plot(depths, avgSizeArrR, label = 'Research')
plt.title('Depth-Size')
plt.xlabel('Depth')
plt.ylabel('Size')
plt.legend()
plt.show()
```

```python
# To test the hypothesis, we compare our research with the original decision tree. The binary tree produces lower accuracy compared to our original decision tree because each node has only two leaf nodes, so the tree grows much slower compared to our original decision tree. We expect two trees to produce the same result. As you can see in the first graph of depth vs. accuracy, the two lines eventually merge at depth 8. The size of the tree is smaller compared the original decision tree because we only extend two children nodes from a parent node. 

# By using voting dataset as the example, it actually does not test the ability of the binary tree well because we only have three attribute values -1, 0, 1 for each attribute. Due to the small number of attribute values, the algorithm will only be picking between two cutoff line. We did not choose to test spam or volacanoes because most of the attributes are continuous, so most nodes are already binary. I think the hypothesis can be better tested if we have mostly nominal attribute type with many attribute values within a dataset.
```

```python
# Research Hypothesis part 2: We find that the log calculations are very expensive in the program, and since log() is a convex function between [0,1], there is a way to simplify the log calculation from O(mn) to O(m). By applying maclaurin equation, we can simplify the log calculation to products. In this way, we can get rid of all log calculations, which will boost the efficiency.

from IPython.display import display, Math, Latex

print('X is attribute and Y is class label. xi is the i-th attribute, yj is the j-th class label. num(x) function means how many different x values.')
display(Math(r'\; \; \; \; \; \; \; \; \; \;\; \; IG(Y|X)=-\sum_{i}\frac{x_{i}}{num(x)}\sum_{j}p(y_{j}|x_{i})\log_2 p(y_{j}|x_{i})'))
display(Math(r'There\:is\:a\:propety\:of\:convex\:function:\;\exists\lambda _{i}\geq 0,\sum_{i=1}^{m}\lambda _{i}=1,\:then\;\sum_{i=1}^{m}\lambda _{i}\cdot f(x_{i})\leq f(\sum_{i=1}^{m}\lambda _{i}\cdot x_{i})'))

print('Since log2 is a convex function on [0,1], p(yj|xi) is greater or equal to 0, and the sum of p(yj|xi) by yj is one, then  we can take out the log2 from the summation. Thus, we have the equation below.')
display(Math(r'\; \; \; \; \; \; \; \; \; \;\; \; IG(Y|X)\geq-\sum_{i}\frac{x_{i}}{num(x)}\log_2(\sum_{j}p(y_{j}|x_{i})^{2})'))
display(Math(r'\; \; \; \; \; \; \; \; \; \;\; \; \;\; \; =-\sum_{i}\frac{x_{i}}{num(x)}log_2\left [{\frac{1}{num(x_{i})^{2}}}\sum_{j}num(y_{j},x_{i})^{2}\right ]'))
display(Math(r'\; \; \; \; \; \; \; \; \; \;\; \; \;\; \; =-\sum_{i}\frac{x_{i}}{num(x)}log_2\left \{ {\frac{1}{num(x_{i})^{2}}}\left [num(y_{1},x_{i})^{2}+num(y_{2},x_{i})^{2}+...+num(y_{j},x_{i})^{2} \right ]\right \}'))
display(Math(r'\; \; \; \; \; \; \; \; \; \;\; \; \;\; \; =-\sum_{i}\frac{x_{i}}{num(x)}log_2\left \{ {\frac{1}{num(x_{i})^{2}}}\left [(num(y_{1},x_{i})+...+num(y_{j},x_{i}))^{2}-(num(y_{1},x_{i})\cdot num(y_{2},x_{i})+...+num(y_{1},x_{i})\cdot num(y_{j},x_{i})+...+num(y_{j-1},x_{i})\cdot num(y_{j},x_{i})) \right ]\right \}'))
display(Math(r'\; \; \; \; \; \; \; \; \; \;\; \; \;\; \; =-\sum_{i}\frac{x_{i}}{num(x)}log_2\left \{ {\frac{1}{num(x_{i})^{2}}}\left [\left [\sum_{j}num(y_{j},x_{i})\right ]^{2}-\sum_{j}\sum_{k, k\neq j}num(y_{j},x_{i})\cdot num(y_{k},x_{i}) \right ]\right \}'))
display(Math(r'\; \; \; \; \; \; \; \; \; \;\; \; \;\; \; =-\sum_{i}\frac{x_{i}}{num(x)}log_2\left \{{\frac{num(x_{i})^{2}}{num(x_{i})^{2}}}-\sum_{j}\sum_{k, k\neq j}\frac{num(y_{j},x_{i})}{num(x_{i})}\cdot \frac{num(y_{k},x_{i})}{num(x_{i})}\right \}'))
display(Math(r'\; \; \; \; \; \; \; \; \; \;\; \; \;\; \; =-\sum_{i}\frac{x_{i}}{num(x)}log_2\left [1-\sum_{j}\sum_{k, k\neq j}p(y_{j},x_{i})\cdot p(y_{k},x_{i})\right ]'))

print('Since y represents class label, where the value of y can only be 0 and 1, so we have')
display(Math(r'\; \; \; \; \; \; \; \; \; \;\; \; \;\; \; =-\sum_{i}\frac{x_{i}}{num(x)}log_2\left [ 1- 2\cdot p(y_{0},x_{i})\cdot p(y_{1},x_{i})\right ]'))

print('According to a special case in Taylor series called Maclaurin polynomial, where')
display(Math(r'\; \; \; \; \; \; \; \; \; \;\; \; \ln{1-x}=-x \; when\:x\rightarrow 0'))

print('And num(yi, xi)num(yk, xi) will approach 0, so we have ')
display(Math(r'\; \; \; \; \; \; \; \; \; \;\; \; log_2\left [ 1- 2\cdot num(y_{0},x_{i})\cdot num(y_{1},x_{i})\right ]=\frac{1}{\ln2}\cdot \ln\left [ 1- 2\cdot num(y_{0},x_{i})\cdot num(y_{1},x_{i})\right ]=- \frac{2}{\ln2}\cdot num(y_{0},x_{i})\cdot num(y_{1},x_{i})'))

print('We have')
display(Math(r'\; \; \; \; \; \; \; \; \; \;\; \; \;\; \; =-\sum_{i}\frac{x_{i}}{num(x)}\left [-\frac{2}{\ln{2}}\cdot \left [p(y_{0},x_{i})\cdot p(y_{1},x_{i})\right ]\right ]'))
display(Math(r'\; \; \; \; \; \; \; \; \; \;\; \; \;\; \; =\frac{2}{\ln{2}}\cdot \sum_{i}\frac{x_{i}}{num(x)}\left [p(y_{0},x_{i})\cdot p(y_{1},x_{i})\right ]'))

print('Since we use IG as comparing metrics, the constant of the equation does not matter. Therefore, we have')
display(Math(r'\; \; \; \; \; \; \; \; \; \;\; \; IG(Y|X)=\sum_{i}\frac{x_{i}}{num(x)}\left [p(y_{0},x_{i})\cdot p(y_{1},x_{i})\right ]'))
```

```python
# Continue: Research Hypothesis part 2

from mainO import getAccuracyAndTime as getO
from mainR import getAccuracyAndTime as getR
from matplotlib import pyplot as plt

avgTimeArrO = []      # Origial avgTime
avgAccuracyArrO = []      # Origial avgSizAccracy
avgSizeArrO = []      # Origial avgTime
avgSizeArrR = []      # Research avgTime
avgTimeArrR = []      # Research avgSize
avgAccuracyArrR = []      # Research avgAccracy
depths = [1, 2, 4, 8, 16, 32]
for depth in depths:
    avgTime, avgAccuracy, avgSize = getO('/volcanoes', 1, depth, 0)
    avgTimeArrO.append(avgTime)
    avgAccuracyArrO.append(avgAccuracy)
    avgSizeArrO.append(avgSize)
    avgTime, avgAccuracy, avgSize = getR('/volcanoes', 1, depth, 0)
    avgTimeArrR.append(avgTime)
    avgAccuracyArrR.append(avgAccuracy)
    avgSizeArrR.append(avgSize)

plt.plot(depths, avgAccuracyArrO, linestyle='dashed', label = 'Original')
plt.plot(depths, avgAccuracyArrR, label = 'Research')
plt.title('Depth-Accuracy')
plt.xlabel('Depth')
plt.ylabel('Accuracy')
plt.legend()
plt.show()

plt.plot(depths, avgTimeArrO, linestyle='dashed', label = 'Original')
plt.plot(depths, avgTimeArrR, label = 'Research')
plt.title('Depth-Time')
plt.xlabel('Depth')
plt.ylabel('Time')
plt.legend()
plt.show()

plt.plot(depths, avgSizeArrO, linestyle='dashed', label = 'Original')
plt.plot(depths, avgSizeArrR, label = 'Research')
plt.title('Depth-Size')
plt.xlabel('Depth')
plt.ylabel('Size')
plt.legend()
plt.show()
```

```python
# The result is suprising. Although we expect the running will decrease, the fact is there is a higher running time with a higher accuracy. This may cause by the inequality scaling of the simplication calculation. While using inequality scaling, the value of IG is changed unbalancedly. As a result, the structure of tree is changed. The new tree has more nodes than the old ones. This increase in size overcomes the simplification in calculation. It shows that although size is much larger than the origianl, the simplified alogrithm still keeps a similar running time when depth is low (between 1 and 5). And when depth is larger, the running time compared to its size is still lower than the origianl algorithm. Although the result of increased running time is not intended, the improvement of accuracy is desired.
```

```python
# Research Hypothesis part 3: We find that with the increasing depth, at certain value, the accuracy will drop off because the program start overfitting. One way to solve it is to use iterative-deepening. With the growth in the depth, if it find there is a decrease in accuracy, the program will stop at that depth and stop future overfitting.

from matplotlib import pyplot as plt
from mainR import getAccuracyAndTime as getR
depths = [1, 4, 8, 12, 16, 20]
AccuracyO = []      # Origial Accuracy
AccuracyR = []      # Research Accuracy
for depth in depths:
    avgTime, avgAccuracy, avgSize = getR('/volcanoes', 1, depth, 0)
    AccuracyO.append(avgAccuracy)
    avgTime, avgAccuracy, avgSize = getR('/volcanoes', 1, -1, 0)      # depth=-1 means it uses iterative-deepening method
    AccuracyR.append(avgAccuracy)

plt.plot(depths, AccuracyO, label = 'Original')
plt.plot(depths, AccuracyR, label = 'Research')
plt.title('Depth-Accuracy')
plt.xlabel('Depth')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
# So, the iterative-deepening decide to stop depth=12 because it shows the accuracy won't increase by the depth anymore from 12. And the original algorithm also shows when depth=12, the accuracy is highest. This proves our hypothesis that use iterative-deepening can prevent over-calculation and even overfitting.
```
