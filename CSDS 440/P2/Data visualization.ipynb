# PROBLEM B
When we add a penalty term to the loss function, the negative conditional log likelihood in our case, we expect differences during minimization. Under two extreme cases, where lambda is equals to 0 and 1, the minimization will behave differently. When lambda is equal to 0, there will be no effect on the penalty term; therefore, we are only minimizing the loss function, and the weight can be very large. When lambda is equal to 1, there will be a heavier focus on the penalty term where we are minimizing both the weight itself and loss function, which can prevent overfitting.
Therefore, we are expecting three results. First is there is no significant different between no penalty and regular penalty. Second is there is an improvement of accuracy because of the success of overfitting control. Third is there is a lower accuracy because the weight of best attributes are decreased.

```python
import numpy as np
from logregNB import Logregression
from matplotlib import pyplot as plt

lambdaList = [0, 0.01, 0.1, 1]

for _lambda in lambdaList:
    accuracy, accuracySD, precision, precisionSD, recall, recallSD, areaUnderROC, FPRs, TPRs = Logregression('/voting', 1, _lambda)
    plt.plot(FPRs, TPRs, label=_lambda)

plt.title('the effect of lambda on voting')
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.legend()
plt.show()
```

For voting data, there is no much difference from lambda = 0 to lambda = 1. When the loss function we are trying to minimize is close to 0, there will be only a small difference between the graph with penalty term and the graph without penalty term because the penalty term reduces the focus on weight. 

```python
import numpy as np
from logregNB import Logregression
from matplotlib import pyplot as plt

lambdaList = [0, 0.01, 0.1, 1]

for _lambda in lambdaList:
    accuracy, accuracySD, precision, precisionSD, recall, recallSD, areaUnderROC, FPRs, TPRs = Logregression('/volcanoes', 0, _lambda)
    plt.plot(FPRs, TPRs, label=_lambda)

plt.title('the effect of lambda on volcanoes')
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.legend()
plt.show()
```

For volcanoes data, the AUC for ROC is the highest when lambda = 0 while AUC for ROC is the lowest when lambda = 1. This can happen if one attribute has larger weight and this attribute is strongly related to classlabel. In this case, volcanoes has a good attribute image_id for classifying the classes. As a result of having penalty term, we reduce the weight of image_id, which cause of higher frequency of errors.

```python
import numpy as np
from logregNB import Logregression
from matplotlib import pyplot as plt

lambdaList = [0, 0.01, 0.1, 1]

for _lambda in lambdaList:
    accuracy, accuracySD, precision, precisionSD, recall, recallSD, areaUnderROC, FPRs, TPRs = Logregression('/spam', 0, _lambda)
    plt.plot(FPRs, TPRs, label=_lambda)

plt.title('the effect of lambda on spam')
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.legend()
plt.show()
```

For spam data, the AUC for ROC is the highest when lambda is 1 while AUC for ROC is the lowest when lambda is 0. By using the penalty term, we can control overfitting by preventing large weights on some attributes. As a result, the accuracy of the model will be improved. This is what we expected when we apply the penalty term.

We have two Research Parts. 

Research Part I is researching on the impact of feature selection on the naive bayes and logistic regression model. ## # Research Part I
- Since naive bayes model uses independence assumption to classify examples, the assumption may not be always correct. To control overfitting, we want to select features that are the most relevant to the class label. As for logistic regression, the discriminative model can reduce the chance of having a few bad attribute that influences the classfication, but it does not hurt if we do a feature selection.
- We know that chi square test is a statistical test applied to sets of categorical data to evaluate the association between two variables. By applying the chi square test to one of the attribute and class label, we are able to find the association of that attribute to the class label.
- We first apply chi square test to all of the attributes and find the highest k scores. We then select the attributes with the highest k scores to run the two different models.
- As we find out in the naive bayes model, feature selection works great. For voting and volcanoes, there are only a few attribute that dominates the classification while the rest has little to none influence to the classfication. By only selecting the k relevant attributes to the class label, we are able to control overfitting of generative model. For voting, we can increase the accuracy by 0.006. For volcano, we can increase the accuracy by 0.07. For spam data, almost half of the attributes are in use for classification while the other half have less influence. Compare to running full attributes, we can improve the accuracy by 0.002-0.003. Therefore, proving that feature selection is important in generative model such as naive bayes.
- After applying the feature selection to logistic regression, we get the result that we expect where feature selection does not help much with classification because logistic regression model does not assume the attributes to be independent. For both voting and volcano data, the accuracy is the highest when we do full attributes. However, for spam, we actually get a slight increase in accuracy when k = 11 by 0.01. Therefore, discriminative model often don't need feature selection to control overfitting unless there are too many noises in the data.

```python
# RESEARCH PART I

# Naive Bayes

from nbayesR import NaiveBayes
from matplotlib import pyplot as plt
import warnings
warnings.filterwarnings('ignore')

kList = list(range(1, 11))
accuracyList = list()

for k in kList:
    accuracy, accuracySD, precision, precisionSD, recall, recallSD, areaUnderROC = NaiveBayes('/voting', 1, 10, 10, k)
    accuracyList.append(accuracy)
plt.plot(kList, accuracyList)
plt.title('The impact of feature selection on Naive bayes accuracy on voting')
plt.xlabel('num of selected feature')
plt.ylabel('accuracy')
plt.show()

kList = list(range(1, 226, 10))
accuracyList = list()

for k in kList:
    accuracy, accuracySD, precision, precisionSD, recall, recallSD, areaUnderROC = NaiveBayes('/volcanoes', 1, 10, 10, k)
    accuracyList.append(accuracy)
plt.plot(kList, accuracyList)
plt.title('The impact of feature selection on Naive bayes accuracy on volcanoes')
plt.xlabel('num of selected feature')
plt.ylabel('accuracy')
plt.show()

kList = list(range(1, 19, 2))
accuracyList = list()

for k in kList:
    accuracy, accuracySD, precision, precisionSD, recall, recallSD, areaUnderROC = NaiveBayes('/spam', 1, 10, 10, k)
    accuracyList.append(accuracy)
plt.plot(kList, accuracyList)
plt.title('The impact of feature selection on Naive bayes accuracy on spam')
plt.xlabel('num of selected feature')
plt.ylabel('accuracy')
plt.show()
```

## # RESEARCH PART II

- Since we use learning rate (LR) to control the step size of gradient descent, we want to modify the LR to reduce the converge time that helps us to reach to minimum. We suggest four different learning rate model:
    * Non-changeable LR (original) 
        * initial LR=0.1
        * LR=0.1
    * Linear-decay LR
        * initial LR=1
        * LR -= 0.0002        
    * Time-based decay LR
        * initial LR=1
        * LR *= (1. / (1. + 0.5 * count_itr))
            * count_itr is the number of iterations
    * Exponential LR
        * initial LR=1
        * LR = np.exp(- 0.1 * count_itr)
            * count_itr is the number of iterations
- We want to research on how different learning rate model affects running time and accuracy. As the bar chart shows below, we draw out the running time as LR type and accuracy as LR type. We find that in logreg, the best learning rate type is linear-decay LR which decay from 1 by 0.0001 each turn. The time-based decay LR has half running time and a bit less accuracy, which will also be a good choice for a large dataset.
- As a conclusion, we find out different learning rate will increase or decrease running time a lot, but will not increase the accuracy, which is controlled by converge rate. At the same time, the shorter running time allows to have much smaller converge change limitation, which may improve accuracy.

```python
# RESEARCH PART II

from logregR2 import Logregression
from matplotlib import pyplot as plt
import warnings
warnings.filterwarnings('ignore')
timeList = list()
accuracyList = list()
for lrType in range(4):
    accuracy, accuracySD, precision, precisionSD, recall, recallSD, areaUnderROC, time = Logregression('/volcanoes', 1, 1, lrType)
    timeList.append(time)
    accuracyList.append(accuracy)

name = ['Non-changeable', 'Linear', 'Time-Based', 'Exponential']
plt.bar(name, timeList)
plt.xlabel("Learning Rate Type")
plt.ylabel("Time")
plt.title("Running Time as learning rate type")
plt.show()

name = ['Non-changeable', 'Linear', 'Time-Based Decay', 'Exponential Decay']
plt.bar(name, accuracyList)
plt.xlabel("Learning Rate Type")
plt.ylabel("Accuracy")
plt.title("Accuracy as learning rate type")
plt.show()
```
